**step 6 : use NearestNeighbors model**

```python
from sklearn.neighbors import NearestNeighbors

from sklearn.metrics.pairwise import cosine_similarity
```

```python
model=NearestNeighbors(n_neighbors=6, metric='euclidean')
model.fit(x_train,y_train)
```

```python
def rec(movie_title):
    if movie_title not in df['title'].values:
        return "Movie not found in the dataset. Please enter a valid movie title."

    movie_idx = df[df['title'] == movie_title].index[0]


    query_point = x.loc[[movie_idx]]


    distances, indices = model.kneighbors(query_point)


    recommended_movie_titles = df.iloc[indices[0][1:]]["title"]
    return recommended_movie_titles

# Test it
movie = input("enter the movie name: ")
print(rec(movie))
```
